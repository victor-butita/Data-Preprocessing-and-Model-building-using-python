```python
#above notice that some values are range
#they have to be converted to one value by taking average
#sample function that i will try to convert a value in total_sqft column to float
#If its not a valid value it goes to except block
def is_float(x):
  try:
    float(x)
  except:
      return False
  return True
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens) == 2:
    return(float(tokens[0] ) + float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(10)
```

```python
df5 = df4.copy()
df5['price_per_sqft'] = df5['price']*100000/ df5['total_sqft']
df5.head()
```

```python
#explore location column
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip())
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
#locations with less than 10 data points to be referred to as other locations
len(location_stats[location_stats<=10])
```

```python
len(df5.location.unique())
```

```python
#locations with less than 10 data points to be referred to as other locations
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df5.location.unique())
```

```python
df5.head(10)
```

```python
#Outlier detection and Removal
#Standard Dev, Simple domain knowledge

df5[df5.total_sqft/df5.bhk <300].head()
```

```python
df5.shape
```

```python
df6 = df5[~(df5.total_sqft/df5.bhk <300)]
df6.shape
```

```python
#checking price per sqft
df6.price_per_sqft.describe()
```

```python
#function to remove extreme price values based on standard deviation
def remove_pps_outliers(df):
  df_out = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
    df_out = pd.concat([df_out,reduced_df],ignore_index=True)
  return df_out

df7 = remove_pps_outliers(df6)
df7.shape
```

```python
#function to compare property prices
def plot_scatter_chart(df,location):
  bhk2 = df[(df.location==location) & (df.bhk==2)]
  bhk3 = df[(df.location==location) & (df.bhk==3)]
  matplotlib.rcParams['figure.figsize'] = (15,10)
  plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
  plt.scatter(bhk3.total_sqft,bhk3.price,marker='+',color='green',label='3 BHK', s=50)
  plt.xlabel('Total Square Feet Area')
  plt.ylabel('Price Per Square Feet')
  plt.title(location)
  plt.legend()

plot_scatter_chart(df7,'Rajaji Nagar')

```

```python
plot_scatter_chart(df7,'Hebbal')
```

```python
#Observe above that some blue points prices (2bedroom) are higher than the green ones(3 bedroom)
#Lets do a cleanup

#We should remove properties where for same location, he price of 3bdrm apartment is less than 2bdrm apartment
#For a given location, we will build a dictionary of stats per bhk
#Later we can remove those 2bhk apartments whose price_per_sqft is less than mean price_per_sqft of 1bhk

def remove_bhk_outliers(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean':np.mean(bhk_df.price_per_sqft),
          'std':np.std(bhk_df.price_per_sqft),
          'count':bhk_df.shape[0]
      }
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices, axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
df9 = df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
#Now our df is reading for modelling and we can drop the size, price_per_sqft features since they are unnecessary

df10 = df9.drop(['size','price_per_sqft'], axis='columns')
df10.head()
```

```python
#Model building
#Encoding
dummies = pd.get_dummies(df10.location)
dummies.head()
```

```python
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis='columns')
df11.head()
```

```python
df12 = df11.drop('location', axis='columns')
df12.head()
```

```python
df12.shape
```

```python
#Create X variable containing independent variables

X = df12.drop('price',axis='columns')
X.head()
```

```python
#Create the y variable containing the dependent variable
y = df12.price
y.head()
```

```python
# Train dataset = model training
# Test dataset = model performance
# test_size = 0.2, 20% of data to be used for testing while 80% for training

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
#create a linear regression model

from sklearn.linear_model import LinearRegression

# Creating an instance of the Linear Regression model
lr_clf = LinearRegression()

# Training the model with the training data
lr_clf.fit(X_train,y_train)

# Evaluating the model's performance on the test data and obtaining the R-squared score
lr_clf.score(X_test,y_test)
```

```python
#This line imports the ShuffleSplit class, which is a cross-validation strategy that shuffles and splits the dataset into train/test sets for multiple iterations.
from sklearn.model_selection import ShuffleSplit

#This line imports the cross_val_score function, which is used for cross-validation and model evaluation.
from sklearn.model_selection import cross_val_score

#This line creates an instance of the ShuffleSplit cross-validation strategy. It specifies that the dataset will be split into 5 different training/testing sets,
#each with a test size of 20%, and sets a random seed for reproducibility
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

#this line uses the cross_val_score function to perform cross-validation on a linear regression model
cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python


from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearch(X,y):
  algos = {
      'linear_regression': {
          'model':LinearRegression(),
          'params':{
              #'normalize': [True, False]
          }
      },
      'lasso':{
          'model':Lasso(),
          'params': {
              'alpha': [1,2],
              'selection': ['random','cyclic']
          }
      },
      'decision_tree': {
          'model': DecisionTreeRegressor(),
          'params': {
              'criterion' : ['mse','friedman_mse'],
              'splitter': ['best','random']
          }
      }
  }
  scores = []
  cv = ShuffleSplit(n_splits=5,test_size=0.2, random_state=0)
  for algo_name, config in algos.items():
    gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
    gs.fit(X,y)
    scores.append({
        'model': algo_name,
        'best_score': gs.best_score_,
        'best_params':gs.best_params_
    })

  return pd.DataFrame(scores,columns=['model','best_score','best_params'])

```

```python
find_best_model_using_gridsearch(X,y)
```

```python
def predict_price(location,sqft,bath,bhk):
  loc_index = np.where(X.columns == location)[0][0]

  x = np.zeros(len(X.columns))
  x[0] = sqft
  x[1] = bath
  x[2] = bhk
  if loc_index >= 0:
    x[loc_index] = 1

  return lr_clf.predict([x])[0]
```
